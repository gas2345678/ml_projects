```python
from sklearn.svm import SVC
import pandas as pd
import math
```

```python
df = pd.read_csv(r"..\ml_projects\ml_datasets/amazon_analysis.csv")
```

```python
df.head(5)
```

```python
df = df.dropna()
df.head(5)
```

We will extract the features that will be important to predict if a best a seller is considered "Best Seller" or not.

So we have to drop the unnecesary columns.

```python
x = df.drop(columns=["product_title","total_reviews","original_price","discounted_price","is_sponsored","buy_box_availability","has_coupon","delivery_date","product_image_url","product_page_url","product_category","discount_percentage","sustainability_tags","data_collected_at","is_best_seller"])
print(x.head(5))
y = df["is_best_seller"]
print(y.head(5))
```

Create an instance of support vector model

```python
svc_model = SVC(max_iter=100,kernel="sigmoid",verbose=True)
```

Now let's divide our dataset into training data and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

Let's fit the model

```python
svc_model.fit(x_train,y_train)
```

```python
predicted = svc_model.predict(x_test)
print(predicted[0:10])
```

Here we can see the total score of the predictions

```python
total_score = svc_model.score(x_test,y_test)
print("Total score: " + str((total_score * 100)))
```

As we can see, the model has achieved a 93% accuracy level.
